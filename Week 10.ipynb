Activation 중에서 sigmoid, tanh, relu 이 세 종류가 있다. sigmoid는 모든 값들을 0하고 1 사이로 넣어버리는 것. squash function이라고 부르기도 함. squash function에 0하고 1 사이로 squash 하는 function이 있고, -1에서 1 사이로 squash 하는 function도 있다==이게 tanh. relu는 특정값은 0으로 나머지는 그대로. softmax는 총합으로 나눠준다. percent화 시킨다. activation은 non linearity를 부과한다가 기본. 마지막에 sigmoid를 걸어버리면 0-1사이로 나오게 된다.

```python
model = Sequential()
model.add(Dense(10, input_shape=[13], activation='relu'))
model.add(Dense(13, activation='relu'))
model.add(Dense(20))
model.add(Dense(1))
model.compile (optimizer='rmsprop', loss='mse')
model.summary()
```
