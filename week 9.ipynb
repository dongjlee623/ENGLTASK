Dense = fully connected. CNN convolutional neural network. (듬성듬성) neural network 사이에 connection dot들을 node. edge는 연결해주는 선들. adjacent 한 layer 끼리는 connect가 되지만, 멀리 떨어져 잇으면 연결이 되지 않는다. data에 pair로 값들이 있다. pairwise로 데이터가 존재한다. 입력값은 input, 출력값은 target이라고 한다. output쪽에 해당되는 data는 target data라고 한다. weight matrix가 남고, network가 예측할 것을 기대하면서 input을 넣으면 계산되어서 target data가 나온다. 계산값은 output. output은 예측되는 데이터를 output이라 한다. 중간에 들어가있는 node들은 hidden layer라고 한다. x와 y의 관계가 linear다 라는 건 한 선으로 나타내어지는 것. 1과 0으로 category를 나누는 것은 중요. continuous한 값을 categorical로 전환하는 것조차 non linear라고 보면 된다. 입력은 continous하게 들어오더라도 출력부분이 categorical 할 수도 있다. 제일 마지막에다가 y=ax+b에 (ax+b)에 sigunoid를 붙이면 된다.

```python
import tensorflow
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense

import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.boston_housing.load_data(path = 'boston_housing.npz')
```

```python
model = Sequential()
model.add(Dense(1, input_shape=[13]))
model.compile (optimizer='rmsprop', loss='mse')
model.summary()
```

```python
model.fit(x_train, y_train, epochs=1000)
```

```python
model.evaluate(x_test, y_test)
```
